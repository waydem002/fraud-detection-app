# Note: Model Performance Comparison

Both Logistic Regression and Random Forest achieved strong ROC-AUC scores, indicating good ranking ability for fraud detection.

However, key differences emerged when focusing on fraud-class performance, which is the primary business objective.

Logistic Regression (with SMOTE):

Achieved very high recall, successfully identifying most fraudulent transactions

Suffered from extremely low precision, resulting in many false positives

Suitable when recall is the only priority, but costly in real-world scenarios

Random Forest (with SMOTE + threshold tuning):

Maintained strong recall while significantly improving precision

Achieved the highest F1-score for the fraud class

Provided better balance between catching fraud and minimizing false alarms

The Random Forest model with SMOTE and optimized decision threshold was selected as the final model.

Reasons for selection:
